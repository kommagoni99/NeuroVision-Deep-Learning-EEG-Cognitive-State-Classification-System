```python
# ==========================================================
# NEUROVISION: EEG COGNITIVE STATE DEEP LEARNING SYSTEM
# ==========================================================

!pip install numpy pandas matplotlib seaborn scikit-learn tensorflow scipy

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.signal import welch
from scipy.fft import fft
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import LabelEncoder
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, MaxPooling1D, Flatten, LSTM, Dropout
from tensorflow.keras.utils import to_categorical

np.random.seed(42)
tf.random.set_seed(42)

# ----------------------------------------------------------
# 1. EEG SIGNAL SIMULATION
# ----------------------------------------------------------

def generate_eeg(state, length=256, fs=128):
    t = np.linspace(0, 1, length)

    if state == "Relaxed":
        signal = np.sin(2*np.pi*10*t)
    elif state == "Focused":
        signal = np.sin(2*np.pi*20*t)
    elif state == "Stressed":
        signal = np.sin(2*np.pi*35*t)
    elif state == "Drowsy":
        signal = np.sin(2*np.pi*5*t)

    noise = np.random.normal(0, 0.5, length)
    return signal + noise

states = ["Relaxed", "Focused", "Stressed", "Drowsy"]

signals = []
labels = []

for _ in range(1000):
    for state in states:
        signals.append(generate_eeg(state))
        labels.append(state)

signals = np.array(signals)
labels = np.array(labels)

# ----------------------------------------------------------
# 2. FREQUENCY FEATURE EXTRACTION
# ----------------------------------------------------------

def extract_band_power(signal, fs=128):
    freqs, psd = welch(signal, fs)
    delta = np.mean(psd[(freqs >= 0.5) & (freqs <= 4)])
    theta = np.mean(psd[(freqs >= 4) & (freqs <= 8)])
    alpha = np.mean(psd[(freqs >= 8) & (freqs <= 12)])
    beta = np.mean(psd[(freqs >= 13) & (freqs <= 30)])
    gamma = np.mean(psd[(freqs >= 30) & (freqs <= 45)])
    return [delta, theta, alpha, beta, gamma]

features = np.array([extract_band_power(s) for s in signals])

# ----------------------------------------------------------
# 3. ENCODE LABELS
# ----------------------------------------------------------

le = LabelEncoder()
y_encoded = le.fit_transform(labels)
y_cat = to_categorical(y_encoded)

X_train, X_test, y_train, y_test = train_test_split(
    signals, y_cat, test_size=0.2, random_state=42
)

# reshape for CNN/LSTM
X_train = X_train.reshape(-1, 256, 1)
X_test = X_test.reshape(-1, 256, 1)

# ----------------------------------------------------------
# 4. CNN MODEL
# ----------------------------------------------------------

cnn_model = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(256,1)),
    MaxPooling1D(2),
    Conv1D(64, 3, activation='relu'),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(states), activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\nTraining CNN...")
cnn_history = cnn_model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2)

# ----------------------------------------------------------
# 5. LSTM MODEL
# ----------------------------------------------------------

lstm_model = Sequential([
    LSTM(64, input_shape=(256,1)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(len(states), activation='softmax')
])

lstm_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\nTraining LSTM...")
lstm_history = lstm_model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2)

# ----------------------------------------------------------
# 6. HYBRID CNN-LSTM
# ----------------------------------------------------------

hybrid_model = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(256,1)),
    MaxPooling1D(2),
    LSTM(64),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(len(states), activation='softmax')
])

hybrid_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("\nTraining Hybrid CNN-LSTM...")
hybrid_history = hybrid_model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2)

# ----------------------------------------------------------
# 7. EVALUATION FUNCTION
# ----------------------------------------------------------

def evaluate_model(model, name):
    preds = model.predict(X_test)
    pred_classes = np.argmax(preds, axis=1)
    true_classes = np.argmax(y_test, axis=1)

    print(f"\n{name} Classification Report:\n")
    print(classification_report(true_classes, pred_classes, target_names=states))

    cm = confusion_matrix(true_classes, pred_classes)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, cmap="Blues", fmt='d')
    plt.title(f"{name} Confusion Matrix")
    plt.show()

evaluate_model(cnn_model, "CNN")
evaluate_model(lstm_model, "LSTM")
evaluate_model(hybrid_model, "Hybrid")

# ----------------------------------------------------------
# 8. TRAINING CURVES
# ----------------------------------------------------------

plt.figure(figsize=(8,5))
plt.plot(cnn_history.history['accuracy'], label="CNN")
plt.plot(lstm_history.history['accuracy'], label="LSTM")
plt.plot(hybrid_history.history['accuracy'], label="Hybrid")
plt.legend()
plt.title("Model Accuracy Comparison")
plt.show()

# ----------------------------------------------------------
# 9. SAMPLE SIGNAL VISUALIZATION
# ----------------------------------------------------------

sample_signal = generate_eeg("Focused")
plt.figure(figsize=(10,4))
plt.plot(sample_signal)
plt.title("Sample EEG Signal (Focused)")
plt.show()

print("\nNeuroVision system complete.")
```
